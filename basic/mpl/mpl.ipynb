```python
import numpy as np
```

```python
x = np.linspace(0, 10, 50)
y = x
y2 = [i**2 for i in x]
plt.title("  y  x, y2  x^2")
plt.xlabel("x")
plt.ylabel("y1, y2")
plt.grid()
plt.plot(x, y, x, y2)
```

```python
x = np.linspace(0, 10, 50)
y = x
y2 = [i**2 for i in x]

plt.figure(figsize=(9, 9))
plt.subplot(2, 1, 1)
plt.plot(x, y)
plt.title("y(x)")
plt.ylabel('y', fontsize=14)
plt.grid(True)

plt.subplot(2, 1, 2)
plt.plot(x, y2)
plt.title("y(x^2)")
plt.ylabel('y2', fontsize=14)
plt.xlabel('x', fontsize=14)
plt.grid()
```

```python
fruits = ['apple', 'peach', 'banana', 'orange', 'melone']
counts = [10, 20, 15, 5, 25]
plt.bar(fruits, counts)
plt.title("Fruits", fontsize=20)
plt.xlabel("fruits")
plt.ylabel("counts")
```
