Ее можно вычислить, используя только предсказанные вероятности и истинные классы без конкретного порога классификации + она раотает даже если классы в данных сильно несбалансированны (примеров одного класса в десятки раз больше примеров длугого). Именно поэтому она очень удобна для соревнований.

Посчитать ее легко:

```python
from sklearn.linear_model import LogisticRegression, LogisticRegressionCV
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler, RobustScaler, LabelEncoder, OneHotEncoder, MinMaxScaler
from sklearn.pipeline import make_pipeline
from sklearn.compose import make_column_transformer
```

```python
sum(data.isna().values == True)
```

```python
data_cat = data[cat_cols_essential]
scaler = StandardScaler() # создаем объект нормализатора
# scaler = MinMaxScaler() # создаем объект нормализатора
scaler.set_output(transform="pandas")

data_num_scale = scaler.fit_transform(data[num_cols])

trans = make_column_transformer((OneHotEncoder(drop="if_binary", sparse_output=False), cat_cols_essential), remainder="passthrough")
trans_ed = trans.fit_transform(data_cat)
# print(trans_ed)

data_new = pd.DataFrame(trans_ed, columns=trans.get_feature_names_out())

data_new[num_cols] = data_num_scale
```

```python
X = data_new.values
y = data[target_col]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33)
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)

model1 = LogisticRegression(random_state=10, class_weight="balanced", C=100).fit(X, y)
y_pred = model1.predict(X)
```

```python
# произведем оценку модели
roc_auc_score(y_pred, y), accuracy_score(y_pred, y), f1_score(y_pred, y, average='macro')
```

```python
model2 = LogisticRegression(random_state=10, class_weight="balanced").fit(X_train, y_train)
y_pred_test = model2.predict(X_test)
y_pred_train = model2.predict(X_train)

print("На тестовой выборке: ", roc_auc_score(y_pred_test, y_test),accuracy_score(y_pred_test, y_test), f1_score(y_pred_test, y_test, average='macro'))
print("На обучающей выборке: ", roc_auc_score(y_pred_train, y_train), accuracy_score(y_pred_train, y_train), f1_score(y_pred_train, y_train, average='macro'))
```

```python
model3 = LogisticRegression(random_state=10, class_weight="balanced")
params = {
    'C': [100, 10, 1, 0.1, 0.01, 0.001]
}

model3_grid = GridSearchCV(model3, params, cv=5, scoring='roc_auc', n_jobs=-1, refit=True)
model3_grid.fit(X_train, y_train);
```

```python
print("best_param_", model3_grid.best_params_)
print("best_Score_", model3_grid.best_score_)
y_pred_test = model3_grid.predict(X_test)
y_pred_train = model3_grid.predict(X_train)
print("На тестовой выборке: ", roc_auc_score(y_pred_test, y_test),accuracy_score(y_pred_test, y_test), f1_score(y_pred_test, y_test, average='macro'))
print("На обучающей выборке: ", roc_auc_score(y_pred_train, y_train), accuracy_score(y_pred_train, y_train), f1_score(y_pred_train, y_train, average='macro'))
```

```python
best_model = LogisticRegression(random_state=10, class_weight="balanced", C=10).fit(X_train, y_train)
y_pred_train = best_model.predict(X_train)
y_pred_test = best_model.predict(X_test)
print(roc_auc_score(y_pred_train, y_train), accuracy_score(y_pred_train, y_train), f1_score(y_pred_train, y_train))
print(roc_auc_score(y_pred_test, y_test), accuracy_score(y_pred_test, y_test), f1_score(y_pred_test, y_test))
```
